```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
s=Service(r'C:/Users/Nadeem Anwar/Downloads/edgedriver_win64/msedgedriver.exe')
import time
from bs4 import BeautifulSoup

```

```python
driver=webdriver.Edge(service=s)
```

```python
driver.get('https://www.ajio.com/search/?text=bagpacks')
time.sleep(2)
```

```python
old_height_of_page = driver.execute_script("return document.body.scrollHeight")
print(old_height_of_page)
```

- scrolling the page till last

```python
while True:
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    new_height=driver.execute_script("return document.body.scrollHeight")
    
    if new_height==old_height_of_page:
        break
    old_height_of_page=new_height

```

- getting html of the page

```python
html=driver.page_source
```

```python
with open ('ajio.html','w',encoding='utf-8') as f:
    f.write(html)
```

```python
with open('ajio.html', 'r', encoding='utf-8') as file:
    html_data = file.read()
```

```python
soup=BeautifulSoup(html_data)
soup.prettify()
```
