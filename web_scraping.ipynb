```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36 Edg/116.0.1938.54"
}
page=requests.get(url,headers=headers,timeout=10)
soup=BeautifulSoup(page.text,'html.parser')
print(soup.prettify())
```

```python
columns=[]
for column in soup.find_all('th') :
    if column.isdigit():
           break
    else:
       columns.append(column.text.strip())
print(columns)
```

```python
all_columns=[]
for column in soup.find_all('th'):
    if column.text.strip().isdigit():
        break
    else:
        all_columns.append(column.text.strip())
print(all_columns)


```

```python
all_columns=[]
table=soup.find_all("table")[0]
for column in table.find_all("th"):
       all_columns.append(column.text.strip())
print(all_columns)
```

```python
final_columns=['Ranks', 'Name', 'Industry', 'Revenue', 'Profit', 'Employees', 'Headquarters','none']

```

```python

all_data=soup.find_all('tr')
df_data=[]

for row_data in all_data[2:]:
    all_td=row_data.find_all('td')
    ind_row_data=[data.text.strip() if data else " " for data in all_td]
    df_data.append(ind_row_data)
print(df_data)
  

```

```python
df=pd.DataFrame(df_data,columns=final_columns)
print(df)

```
