```python
webpage=requests.get('https://www.flipkart.com/search?q=smartphones',headers={
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36 Edg/116.0.1938.54"
}).text
```

```python
soup=BeautifulSoup(webpage,'html.parser')

```

```python

phones_img_url=[]
phone_name=[]
all_phones_avg_ratings=[]
all_ratings=[]
all_phones_actual_price=[]
all_phones_discounted_price=[]
total_reviews=[]
all_divs=soup.find_all('div',class_='tUxRFH')
for phone in all_divs:
    img=phone.find('img',class_='DByuf4')['src']
    try:
        phones_img_url.append( img)
    except:
        phones_img_url.append(np.nan)

    #name_of_phones=phone.find('div',class_='KzD1HZ')
    #print(name_of_phones)
    ratings=phone.find('div','XQDdHH')
    try:
        all_phones_avg_ratings.append(ratings.text.strip())
    except:
        all_phones_avg_ratings.append(np.nan)


    all_rating_of_ind=phone.find('span',class_='Wphh3N').find('span').find('span').text.strip()
    try:
        all_ratings.append(all_rating_of_ind)
    except:
        all_ratings.append(np.nan)
    
    try:
        actual_price=phone.find('div',class_="yRaY8j ZYYwLA").text.strip()
    except:
        actual_price=np.nan

    try:
        all_phones_actual_price.append(actual_price)
    except:
        all_phones_actual_price.append(np.nan)

    
    dis_price=phone.find('div',class_="Nx9bqj _4b5DiR").text.strip()
    try:
        all_phones_discounted_price.append(dis_price)
    except:
        all_phones_discounted_price.appned(np.nan)

    review=phone.find('span',class_= "Wphh3N").find('span').text.split('&')[1]
    try:
        total_reviews.append(review)
    except:
        total_reviews.append(np.nan)


```

```python
final_df = pd.DataFrame()

for i in range(1,46):
    webpage=requests.get('https://www.flipkart.com/search?q=smartphones{}'.format(i),headers={
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36 Edg/116.0.1938.54"
    }).text
    
    soup=BeautifulSoup(webpage,'html.parser')
    phones_img_url=[]
    phone_name=[]
    all_phones_avg_ratings=[]
    all_ratings=[]
    all_phones_actual_price=[]
    all_phones_discounted_price=[]
    total_reviews=[]
    all_divs=soup.find_all('div',class_='tUxRFH')
    for phone in all_divs:
        img=phone.find('img',class_='DByuf4')['src']
        try:
            phones_img_url.append( img)
        except:
            phones_img_url.append(np.nan)

        #name_of_phones=phone.find('div',class_='KzD1HZ')
        #print(name_of_phones)
        ratings=phone.find('div','XQDdHH')
        try:
            all_phones_avg_ratings.append(ratings.text.strip())
        except:
            all_phones_avg_ratings.append(np.nan)
        
        try:
            all_rating_of_ind=phone.find('span',class_='Wphh3N').find('span').find('span').text.strip()
        except:
            all_rating_of_ind=np.nan

        try:
            all_ratings.append(all_rating_of_ind)
        except:
            all_ratings.append(np.nan)
        
        try:
            actual_price=phone.find('div',class_="yRaY8j ZYYwLA").text.strip()
        except:
            actual_price=np.nan

        try:
            all_phones_actual_price.append(actual_price)
        except:
            all_phones_actual_price.append(np.nan)

        try:
           dis_price=phone.find('div',class_="Nx9bqj _4b5DiR").text.strip()
        except:
            dis_price=np.nan
        try:
            all_phones_discounted_price.append(dis_price)
        except:
            all_phones_discounted_price.appned(np.nan)
        try:
            review=phone.find('span',class_= "Wphh3N").find('span').text.split('&')[1]
        except:
            review=np.nan
        try:
            total_reviews.append(review)
        except:
            total_reviews.append(np.nan)

        df=pd.DataFrame({'avg_ratings':all_phones_avg_ratings,'all_ratings':all_ratings,'actual_price':all_phones_actual_price,'with_discount':all_phones_discounted_price,'reviews':total_reviews})
        final_df=pd.concat([final_df,df],ignore_index=True) 
        all_phones_avg_ratings.clear()
        all_ratings.clear()
        all_phones_actual_price.clear()
        all_phones_discounted_price.clear()
        total_reviews.clear()
```

```python
final_df.to_csv('flipkart.csv')
```
